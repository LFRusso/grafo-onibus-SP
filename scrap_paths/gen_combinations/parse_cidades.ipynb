```python
import numpy as np
import pandas as pd
from unicodedata import normalize
```

```python
# Remove os acentos de uma string
def remover_acentos(txt):
    return normalize('NFKD', txt).encode('ASCII', 'ignore').decode('ASCII')
```

```python
siglas_uf = {
    "Roraima": "RR",
    "Amazonas": "AM",
    "Acre": "AC",
    "Rondônia": "RO",
    "Alagoas": "AL",
    "Amapá": "AP",
    "Bahia": "BA",
    "Ceará": "CE",
    "Espírito Santo": "ES",
    "Goiás": "GO",
    "Maranhão": "MA",
    "Mato Grosso": "MT",
    "Mato Grosso do Sul": "MS",
    "Minas Gerais": "MG",
    "Pará": "PA",
    "Paraíba": "PB",
    "Paraná": "PR",
    "Pernambuco": "PE",
    "Piauí": "PI",
    "Rio de Janeiro": "RJ",
    "Rio Grande do Norte": "RN",
    "Rio Grande do Sul": "RS",
    "Santa Catarina": "SC",
    "São Paulo": "SP",
    "Sergipe": "SE",
    "Tocantins": "TO",
    "Distrito Federal": "DF"
    }
```

```python
df = pd.read_excel("./RELATORIO_DTB_BRASIL_MUNICIPIO.xls")
df = df.drop(['Código Município Completo', 'Mesorregião Geográfica', 'Nome_Microrregião'], axis=1)
print(df.shape)
df.head(20)
```

```python
# Selecionando todas as cidaes e seus estados
cidades = [[remover_acentos(cidade[0].lower().replace(" ", "-").replace("'", "")), cidade[1]] 
           for cidade in np.array(df[['Nome_Município', 'Nome_UF']])]

print(len(cidades))
cidades[:10]
```

```python
# Obtendo cidades do estado de São Paulo
sp = df[['Nome_UF', 'Nome_Município']].query("Nome_UF == 'São Paulo'")
cidades_sp = [remover_acentos(cidade[0].lower().replace(' ', '-')) for cidade in np.array(sp[['Nome_Município']])]
len(cidades_sp)
cidades_sp[:10]
```

```python
## Obtendo todas as combinações entre cidades de SP e salvando na formatação desejada para o scrap
caminhos_sp=[]
for cidade1 in cidades_sp:
    for cidade2 in cidades_sp:
        if(cidade1 != cidade2):
            caminhos_sp.append("{}-sp-br-x-{}-sp-br".format(cidade1, cidade2))

print(len(caminhos_sp))
caminhos_sp[:10]
```
